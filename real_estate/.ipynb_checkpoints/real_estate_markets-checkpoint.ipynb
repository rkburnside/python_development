# data parsing for top 100 markets
- this script will take a whole bunch of <xml> tags and line them up into columns

original website: https://www.buildium.com/blog/up-and-coming-real-estate-markets-2020/

# this is the sudo code
1. pull in a line
2. check: new section
  1. true: start collecting data
  2. false: move to next line
3. check if this is the end of the information
  1. true: move to next line
  2. false: test, collect information, and set variable
3. repeat, set the variable and return nothing


```python
# xml tag variables
tags = {
"a" : "<begin>", 
"b" : "<1-year projected home value growth>", 
"c" : "<5-year projected employment growth>", 
"d" : "<5-year projected population growth>", 
"e" : "<Cost of doing business index>", 
"f" : "<Cost of living index>", 
"g" : "<Home affordability index>", 
"h" : "<Median 1-bedroom rent>", 
"i" : "<Median 2-bedroom rent>", 
"j" : "<Median home value>", 
"k" : "<Median list price per square foot>", 
"l" : "<Metropolitan statistical area>", 
"m" : "<Percentage of residents who rent their homes>", 
"n" : "<Population growth since 2010>", 
"o" : "<Population growth since 2017>", 
"p" : "<Population>", 
"q" : "<Ranking>", 
"r" : "<Rent as % of household income>", 
"s" : "<Rent list price per square foot>", 
"t" : "<Rent to cost of ownership ratio>", 
"u" : "<Rental breakeven horizon>", 
"v" : "<Vacancy rate>", 
"w" : "<Year-over-year home value growth>", 
"x" : "<Year-over-year rent growth>",
"y" : "<end>"
}

# initialize tag values
tag_values = {"a": "", "b": "", "c": "", "d": "", "e": "", "f": "", "g": "", "h": "", "i": "", "j": "", "k": "", "l": "","m": "", "n": "", "o": "", "p": "", "q": "", "r": "", "s": "", "t": "", "u": "", "v": "", "w": "", "x": "", "y": ""}
```

```python
def check_value(current_line, tags, tag_values):
    if tags['a'] == current_line[0]:  # <begin>
        tag_values['a'] = current_line[0]
        return
    elif tags['b'] == current_line[0]:  # <1-year projected home value growth>
        tag_values['b'] = current_line[1]
        return
    elif tags['c'] == current_line[0]:  # <5-year projected employment growth>
        tag_values['c'] = current_line[1]
        return
    elif tags['d'] == current_line[0]:  # <5-year projected population growth>
        tag_values['d'] = current_line[1]
        return
    elif tags['e'] == current_line[0]:  # <Cost of doing business index>
        tag_values['e'] = current_line[1]
        return
    elif tags['f'] == current_line[0]:  # <Cost of living index>
        tag_values['f'] = current_line[1]
        return
    elif tags['g'] == current_line[0]:  # <Home affordability index>
        tag_values['g'] = current_line[1]
        return
    elif tags['h'] == current_line[0]:  # <Median 1-bedroom rent>
        tag_values['h'] = current_line[1]
        return
    elif tags['i'] == current_line[0]:  # <Median 2-bedroom rent>
        tag_values['i'] = current_line[1]
        return
    elif tags['j'] == current_line[0]:  # <Median home value>
        tag_values['j'] = current_line[1]
        return
    elif tags['k'] == current_line[0]:  # <Median list price per square foot>
        tag_values['k'] = current_line[1]
        return
    elif tags['l'] == current_line[0]:  # <Metropolitan statistical area>
        tag_values['l'] = current_line[1]
        return
    elif tags['m'] == current_line[0]:  # <Percentage of residents who rent their homes>
        tag_values['m'] = current_line[1]
        return
    elif tags['n'] == current_line[0]:  # <Population growth since 2010>
        tag_values['n'] = current_line[1]
        return
    elif tags['o'] == current_line[0]:  # <Population growth since 2017>
        tag_values['o'] = current_line[1]
        return
    elif tags['p'] == current_line[0]:  # <Population>
        tag_values['p'] = current_line[1]
        return
    elif tags['q'] == current_line[0]:  # <Ranking>
        temp = current_line[1].split('#')
        temp = temp[1].split(':')
        tag_values['q'] = temp[0]
        return
    elif tags['r'] == current_line[0]:  # <Rent as % of household income>
        tag_values['r'] = current_line[1]
        return
    elif tags['s'] == current_line[0]:  # <Rent list price per square foot>
        tag_values['s'] = current_line[1]
        return
    elif tags['t'] == current_line[0]:  # <Rent to cost of ownership ratio>
        tag_values['t'] = current_line[1]
        return
    elif tags['u'] == current_line[0]:  # <Rental breakeven horizon>
        tag_values['u'] = current_line[1]
        return
    elif tags['v'] == current_line[0]:  # <Vacancy rate>
        tag_values['v'] = current_line[1]
        return
    elif tags['w'] == current_line[0]:  # <Year-over-year home value growth>
        tag_values['w'] = current_line[1]
        return
    elif tags['x'] == current_line[0]:  # <Year-over-year rent growth>
        tag_values['x'] = current_line[1]
        return
    elif tags['y'] == current_line[0]:  # <end>
        tag_values['y'] = current_line[0]
        return
    else:
        return 'n/a'
```

```python
print(tags['a'] + '|' + tags['b'] + '|' + tags['c'] + '|' + tags['d'] + '|' + tags['e'] + '|' + tags['f'] + '|' + tags['g'] + '|' + tags['h'] + '|' + tags['i'] + '|' + tags['j'] + '|' + tags['k'] + '|' + tags['l'] + '|' + tags['m'] + '|' + tags['n'] + '|' + tags['o'] + '|' + tags['p'] + '|' + tags['q'] + '|' + tags['r'] + '|' + tags['s'] + '|' + tags['t'] + '|' + tags['u'] + '|' + tags['v'] + '|' + tags['w'] + '|' + tags['x'] + '|' + tags['y'])
filename = 'test_data.txt'
with open(filename) as data:
    for line in data: # this will iterate line by line through the file
        if line[0:5] == '<end>':
            print(tag_values['a'] + '|' + tag_values['b'] + '|' + tag_values['c'] + '|' + tag_values['d'] + '|' + tag_values['e'] + '|' + tag_values['f'] + '|' + tag_values['g'] + '|' + tag_values['h'] + '|' + tag_values['i'] + '|' + tag_values['j'] + '|' + tag_values['k'] + '|' + tag_values['l'] + '|' + tag_values['m'] + '|' + tag_values['n'] + '|' + tag_values['o'] + '|' + tag_values['p'] + '|' + tag_values['q'] + '|' + tag_values['r'] + '|' + tag_values['s'] + '|' + tag_values['t'] + '|' + tag_values['u'] + '|' + tag_values['v'] + '|' + tag_values['w'] + '|' + tag_values['x'] + '|' + tag_values['y'])
            tag_values = {"a": "", "b": "", "c": "", "d": "", "e": "", "f": "", "g": "", "h": "", "i": "", "j": "", "k": "", "l": "","m": "", "n": "", "o": "", "p": "", "q": "", "r": "", "s": "", "t": "", "u": "", "v": "", "w": "", "x": "", "y": ""}
        elif line[0] == '<':
            temp = line.rstrip('\n')
            temp = temp.split('>')
            temp[0] = temp[0] + '>'
            check_value(temp, tags, tag_values)
```
